1. É um conjunto de dados retirados de anúncios de carros de um jornal, e o principal objetivo é determinar o tempo que o carro ficará anunciado.

```python
# Ler os dados
df = pd.read_csv('cars.csv')

#imprimir as primeiras colunas
df.head()
```

```python
## vamos imprimir os tipos de dados que estã contidos em cada coluna
df.dtypes
```

```python
## podemor ver que há muitas colunas que não são numéricas

## a primeiro alternativa é o chamado OneHotEncoder
## vamos usar a função do pandas 
### https://pandas.pydata.org/docs/reference/api/pandas.get_dummies.html

#vamos fazer o teste com a coluna de cores?
df2 = pd.get_dummies(df, columns = ['color'])
df2.head()
```

```python
## uma outra forma de resolver esse problema é criando uma nova coluna de frequẽncia de ocorrência de um termo:

## vamos calcular as frequências que cada termo aparece:
fq = df.groupby('model_name').size()/len(df)    

# aplicar no map na coluna desejada
df.loc[:, "{}_freq_encode".format('model_name')] = df['model_name'].map(fq)   

df.head(10) 
```

```python
## Além disso podemos criar features binárias
## vamos trabalhar com a coluna engine_fuel

df['engine_fuel_binario'] = np.where(df['engine_fuel'] == 'gasoline', 1, 0)
df.head()
```
