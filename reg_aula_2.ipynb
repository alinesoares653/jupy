```python
#importar as bibliotecas necessárias
import pandas as pd
import numpy as np
```

```python
#ler e imprimir as primeiras linhas
df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
def EDA (df):
    eda_df = {}
    eda_df['Amount_NaN'] = df.isnull().sum()
    eda_df['%_NaN'] = df.isnull().mean().round(2)
    eda_df['DType'] = df.dtypes
    eda_df['Amount_Data'] = df.count()
    
    # Outro ponto para ser verificado, porque para criar a coluna com a quantidade de valores unicos por coluna
    # Não utilizei a função df.unique() 
    colunas = sorted(df.columns.tolist(), key=str.lower, reverse=False)
    eda_df['Amount_Unique'] = filtered_result = list(map(lambda x: len(df[x].unique().tolist()), colunas))
    
    eda_df['Mean'] = np.round(df.mean(), 2)
    eda_df['Median'] = np.round(df.median(), 2)
    
    eda_df['Max'] = df.max()
    eda_df['Min'] = df.min()
    eda_df['STD'] = np.round(df.std(), 2)
    
    return pd.DataFrame(eda_df)
```

```python
#Chamar a função criada acima para gerar o DataFrame
informacao_df = EDA(df)

#imprimir as primeiras linhas
informacao_df
```

```python
# Limpeza inicial

# colunas que chamam a atenção:

# last review - Discussões para simplificar - minha sugestão é deletar - já que não temos uma data de referência

# reviews_per_month - para não perder muitos dados - fazer preechimento

# name - podemos excluir as linhas com dados faltantes

# host_name - podemos excluir as linhas com dados faltantes

# id - elementos unicos 
```

```python
# implementar a limpeza acima
del df['last_review']

df['reviews_per_month'] = df.reviews_per_month.fillna(df.reviews_per_month.mean())

df = df.dropna(subset=['name','host_name'])

del df['id']
```

```python
#conferir se houve mudanças no DataFrame
informacao_df2 = EDA(df)
informacao_df2
```

```python
# depois dessa limpeza inicial vamos continuar o trabalho de simplificação do problema

# Tendo o ponto acima: vamos tirar a coluna name (título do anúncio)
del df['name']

# Considerando as variáveis categóricas:

# host_id e host_name: tem quantidade de valores unicos diferentes (diferentes hosts com mesmo nome)
# se um host chamar Ana, a diária será mais cara, quando comparada com o quarto do João?
del df['host_id']
del df['host_name']
```

```python
#conferir se houve mudanças no DataFrame
informacao_df3 = EDA(df)
informacao_df3
```

```python
# vamos olhar com mais carinho as colunas categóricas remanecentes?
informacao_df3.loc[informacao_df3['DType']=='object']
```

```python
# A grande pergunta é: Como vamos transformar as palavras em números?

# get_dummy: Quais colunas poderiam ser tratadas por esse método?
# frequencia: quais colunas poderiam passar por esse método.

# fazer get_dummy com: neighbourhood_group e room_type

df = pd.get_dummies(df, columns = ['neighbourhood_group', 'room_type'])
df.head()
```

```python
#conferir se houve mudanças no DataFrame
informacao_df4 = EDA(df)
informacao_df4
```

```python
## hora de começar nova etapa: dividir o DataFrame

# vamos usar uma nova classe do sklearn
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
from sklearn.model_selection import train_test_split

#separar o alvo
y = df['price']
del df['price']
```

```python
#fazer a separação
x_treino, x_teste, y_treino, y_teste = train_test_split(df.values, y, test_size=0.3)
```

```python
x_treino_df = x_treino_df.merge(frequencia_df, on='neighbourhood')
x_treino_df.head()
```
